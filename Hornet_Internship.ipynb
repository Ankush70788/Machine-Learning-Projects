```python
import string
!pip install eth_utils
!pip install base58
# Function has been used to extract features
def extract_features(address):
    features = {
        'length': len(address),
        'num_digits': sum(char.isdigit() for char in address),
        'num_uppercase': sum(char.isupper() for char in address),
        'num_lowercase': sum(char.islower() for char in address),
        'num_special': sum(char in string.punctuation for char in address)
    }
    return features

# Apply feature extraction
feature_df = df['Address'].apply(extract_features).apply(pd.Series)

# Concatenate with the original DataFrame
df = pd.concat([df, feature_df], axis=1)
print(df)
#CHECKSUM VALIDATION
!pip install eth-hash[pycryptodome]
import hashlib
import base58
import pandas as pd

# Base58
def base58_decode(address):
    alphabet = '123456789ABCDEFGHJKLMNPQRSTUVWXYZabcdefghijkmnopqrstuvwxyz'
    base58_map = {char: index for index, char in enumerate(alphabet)}

    num = 0
    for char in address:
        num = num * 58 + base58_map[char]

    combined = num.to_bytes(25, byteorder='big')
    return combined[:-4], combined[-4:]

# NEO
def neo_checksum_validation(address):
    try:
        decoded_address, checksum = base58_decode(address)
        hashed = hashlib.sha256(hashlib.sha256(decoded_address).digest()).digest()
        return hashed[:4] == checksum
    except Exception:
        return False

# Bitcoin
def bitcoin_checksum_validation(address):
    try:
        decoded = base58.b58decode_check(address)
        return True
    except ValueError:
        return False

# Ethereum
def ethereum_checksum_validation(address):
    return address.startswith('0x') and len(address) == 42

# Tron
def tron_checksum_validation(address):
    return address.startswith('T') and len(address) == 34

def validate_address(coin, address):
    if coin == 'BTC':
        return bitcoin_checksum_validation(address)
    elif coin == 'ETH':
        return ethereum_checksum_validation(address)
    elif coin == 'TRX':
        return tron_checksum_validation(address)
    elif coin == 'NEO':
        return neo_checksum_validation(address)
    else:
        return False

# Apply checksum validation
df['is_valid'] = df.apply(lambda row: validate_address(row['Coin Type'], row['Address']), axis=1)

# This has been done in order to filter out invalid addresses
df = df[df['is_valid']]

print(df)
```
